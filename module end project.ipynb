```python
# Visualizing employee position in a Barchart
## (Plotting a Bar Chart to visualize the above.)

x= df['Position'].value_counts()
plt.bar(x.index, x)

plt.title('Employee positions at ABC Company')
plt.xlabel('Position')
plt.ylabel('No:of Employees')

plt.show()
```
